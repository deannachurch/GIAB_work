```python
import pandas as pd
```

```python
src_file="CG1781_BaseUpdates.xlsx"
```

```python
df=pd.read_excel(src_file, sheetname=0, convert_float=True)
df
```

```python
df2=df[['#chrom', 'pos', 'id', 'ref', 'alt', 'InAGP']]
df2
```

```python
df3=df2[df2.InAGP=='Y']
df3.columns=['chrom', 'pos', 'id', 'ref', 'alt', 'InAGP']
df3
```

```python
out=open('GRCh37_GRCh38_flipped_sites.bed', 'w')
out.write("track db=GRCh37\n")
for idx in df3.itertuples():
    name = "%s:ref=%s:alt=%s" % (idx.id, idx.ref, idx.alt)
    out.write('%s\t%d\t%d\t%s\n' % (idx.chrom, int(idx.pos)-1, int(idx.pos), name))
out.close()    
```
